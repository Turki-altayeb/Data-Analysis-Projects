```python
# Pandas is used for data handling such as reading Excel and CSV files
import pandas as pd
# Seaborn is used for advanced data visualizations
import seaborn as sns
# Matplotlib is responsible for drawing and displaying or saving plots as images
import matplotlib.pyplot as plt
```

```python
# Load the data from the local file
file_path = r"C:\Users\Mobily\Downloads\heart_disease_uci.csv"
# Opens the file, reads it, and converts it into a data table called DataFrame
df = pd.read_csv(file_path)
```

```python
# Display the first 5 rows
# Used to quickly check the data structure: column names, data types, initial values
print(df.head())

# First, prints a simple sentence: "Dataset Information:" (just for clarity when reading the output)
# Then df.info() shows general information about the data — like saying “give me a summary of the table”, especially types and counts
print("\nDataset Information:")
print(df.info())

# First, prints a descriptive message: "Number of missing values in each column:"
# Then df.isnull() performs a check
# Like saying “show me where I have missing values in the data”
print("\nNumber of missing values in each column:")
print(df.isnull().sum())
```

```python
# This line prepares the drawing canvas
# The goal is to control the plot size so it appears neat and clear
# Like saying “prepare me a page sized 8x5 to draw on”
plt.figure(figsize=(8, 5))
# This uses Seaborn to draw a count plot
# Like saying “count how many people have heart disease and how many don’t, and draw them as bars”
sns.countplot(x='target', data=df)
# Adds a title to the plot
plt.title('Number of Patients by Heart Disease Presence')
# Labels the x-axis
plt.xlabel('Outcome (0 = No Disease / 1 = Has Disease)')
# Labels the y-axis
plt.ylabel('Number of People')
# Displays the final plot on the screen
plt.show()
```

```python
# Shows all the columns in the dataset
print(df.columns)
```

```python
# Open a drawing page with a suitable size before plotting
plt.figure(figsize=(8, 5))
# Uses Seaborn to draw a count plot
# Like saying “I want to draw the number of males and females, and split each by heart disease status”
sns.countplot(x='gender', data=df, hue='target')
# Add a title to the plot for clarity
plt.title('Gender Distribution by Heart Disease Presence')
# Label the x-axis
plt.xlabel('Gender (0 = Female / 1 = Male)')
# Label the y-axis to show that the numbers represent the number of people
plt.ylabel('Number of People')
# Customize the legend to explain what each color means
plt.legend(title='Outcome', labels=['No Disease', 'Has Disease'])
# Display the final plot on screen
```

```python
# Prepare a large, clear canvas
plt.figure(figsize=(10, 6))
# Plot age distribution with separation between diseased and non-diseased
sns.histplot(data=df, x='age', hue='target', multiple='stack', bins=20)
# Add a title to the plot
plt.title('Age Distribution by Heart Disease Presence')
# Label the x-axis (Age)
plt.xlabel('Age')
# Label the y-axis (Number of People)
plt.ylabel('Number of People')
# Show the plot
plt.show()
```
