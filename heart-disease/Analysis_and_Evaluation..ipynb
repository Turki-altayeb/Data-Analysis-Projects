```python
# seaborn for creating nice and organized plots
# matplotlib.pyplot for traditional plots and controlling them
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# read the data
file_path = r"C:\Users\Mobily\Downloads\heart_disease_uci.csv"  # put your file path here
df = pd.read_csv(file_path)

# we set boundaries for age categories (from 0 to 40, from 40 to 55, from 55 to 70, and more than 70 years)
# we gave each category a label (<40, 40-55, 55-70, 70+)
# then we created a new column named age_category that contains the category for each person based on their age
bins = [0, 40, 55, 70, 100]
labels = ['<40', '40-55', '55-70', '70+']
df['age_category'] = pd.cut(df['age'], bins=bins, labels=labels)

# created a new column named age_sex_interaction
# each value = age × disease status (target, if sick = 1, if not = 0)
df['age_sex_interaction'] = df['age'] * df['target']

# create a plot
# x-axis is age categories (age_category)
# hue differentiates between people with disease (target=1) and without (target=0)
# add title, axis labels, legend explaining each color
plt.figure(figsize=(10, 6))
sns.countplot(data=df, x='age_category', hue='target')
plt.title('age distribution by category and heart disease presence')
plt.xlabel('age category')
plt.ylabel('number of people')
plt.legend(title='result', labels=['no disease', 'has disease'])
plt.show()
```

```python
# set figure size (width 10 inches × height 6 inches), to make it clear and large
plt.figure(figsize=(10, 6))
# use seaborn to draw countplot, it counts the number of people
sns.countplot(data=df, x='gender', hue='target')
# add title to the plot, to understand that it shows relationship between gender and heart disease
plt.title('gender distribution and heart disease presence')
# label x-axis as "gender", label y-axis as "number of people"
plt.xlabel('gender')
plt.ylabel('number of people')
# add legend explaining each color in the plot
plt.legend(title='result', labels=['no disease', 'has disease'])
# show the plot
plt.show()
```

```python
# here set figure size (width 10 and height 6) to make it clear and large
plt.figure(figsize=(10, 6))
# use seaborn to draw boxplot
sns.boxplot(data=df, x='target', y='trestbps')
# add title and label axes properly
plt.title('blood pressure level and heart disease presence')
plt.xlabel('heart disease presence')
plt.ylabel('blood pressure')
# change numbers (0 and 1) to readable labels: "no disease" and "has disease"
plt.xticks([0, 1], ['no disease', 'has disease'])
# show the plot
plt.show()
```

```python
# here set figure size (width 10 and height 6) to make data clearer and easier to read
plt.figure(figsize=(10, 6))
# use seaborn to draw boxplot
sns.boxplot(data=df, x='target', y='chol')
# set title and write axis labels in a neat and understandable way
plt.title('cholesterol level and heart disease presence')
plt.xlabel('heart disease presence')
plt.ylabel('cholesterol')
# change numbers 0 and 1 to understandable words
plt.xticks([0, 1], ['no disease', 'has disease'])
# show the plot
plt.show()
```

```python
# set figure size to make it clear (width 10 and height 6)
plt.figure(figsize=(10, 6))
# here we draw bar chart
# each chest pain type (for example: typical angina, atypical angina, non-anginal pain, etc.)
sns.countplot(data=df, x='cp', hue='target')
# write title and set axis labels
plt.title('chest pain type and heart disease presence')
plt.xlabel('chest pain type')
plt.ylabel('number of people')
# explain in legend that one color means "no disease" and another means "has disease"
plt.legend(title='result', labels=['no disease', 'has disease'])
# show the plot
plt.show()
```

```python
# set figure size: width 10 and height 6
plt.figure(figsize=(10, 6))
# boxplot gives you an idea about median (line inside box), data distribution (upper and lower half), outliers
sns.boxplot(data=df, x='target', y='oldpeak')
# change category titles on x-axis
plt.title('exercise-induced ST depression and heart disease presence')
plt.xlabel('heart disease presence')
plt.ylabel('depression level')
# instead of just 0 and 1, show "no disease" and "has disease" in an understandable way
plt.xticks([0, 1], ['no disease', 'has disease'])
# show the plot
plt.show()
```
